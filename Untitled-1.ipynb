Además, se filtraron únicamente los atletas que obtuvieron medalla (oro, plata o bronce), con el objetivo de analizar la distribución y características de los ganadores en estas disciplinas.El conjunto de datos utilizado corresponde a registros históricos de atletas que participaron en los Juegos Olímpicos, incluyendo información demográfica, deportiva y de resultados

1. Carga de datos 

```python
import pandas as pd
df = pd.read_csv("athlete_events.csv")
df.head()
```

2. Filtremos los datos 

Se filtraron únicamente los atletas que obtuvieron medalla (oro, plata o bronce), con el objetivo de analizar la distribución y características de los ganadores en estas disciplinas.

```python
sports = ["Athletics", "Swimming", "Rowing", "Football", "Hockey"]

df_2016 = df[
    (df["Year"] == 2016) &
    (df["Sport"].isin(sports)) &
    (df["Medal"].notna())
]

df_2016.head()
```

3. Veamos las caracteristicas del conjunto de datos.

```python
df.select_dtypes(include=["int64", "float64"]).columns
df.select_dtypes(include=["object"]).columns
print("Variables numéricas:")
print(df.select_dtypes(include=["int64", "float64"]).columns)

print("\nVariables categóricas:")
print(df.select_dtypes(include=["object"]).columns)
```

A partir de la clasificación realizada, se observa que el conjunto de datos contiene una combinación de variables numéricas y categóricas. Las variables numéricas (ID, Age, Height, Weight y Year) y por otro lado, las variables categóricas (como Sex, Team, Sport, Event y Medal).

4. Crea un gráfico utilizando una técnica de visualización adecuada que presente de la mejor manera posible el patrón global de las características de height y weight de los ganadores de medallas de 2016 de los cinco deportes.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df_2016,
    x="Height",
    y="Weight",
    hue="Sport",
    alpha=0.7
)

plt.title("Relación entre Altura y Peso\nGanadores de Medalla - Juegos Olímpicos 2016")
plt.xlabel("Altura (cm)")
plt.ylabel("Peso (kg)")
plt.legend(title="Deporte")
plt.show()
```


- Se observa una relación positivaentre la altura y el peso de los ganadores de medallas en 2016. A medida que aumenta la altura, también tiende a aumentar el peso, lo que indica una correlación directa entre ambas variables.

- Existen diferencias físicas según el deporte. Los atletas de remo tienden a ser más altos y pesados, mientras que natación también presenta atletas altos pero con menor peso promedio. Atletismo muestra mayor dispersión debido a la variedad de disciplinas que incluye. Fútbol y hockey presentan valores más concentrados en rangos intermedios.

- Se aprecia variabilidad dentro de cada deporte, lo que indica que no todos los atletas comparten exactamente las mismas características físicas, aunque sí siguen una tendencia general.

- Existe superposición entre deportes, lo que sugiere que ciertas características físicas pueden ser competitivas en más de una disciplina.

5. Crea un gráfico utilizando una técnica de visualización adecuada que presente de la mejor manera posible la estadística de resumen para la altura y el peso de los jugadores que ganaron cada tipo de medalla (oro/plata/bronce) en los datos.


5.1 Height

```python
import matplotlib.pyplot as plt

# Filtrar solo ganadores de medalla
df_medalists = df[df["Medal"].notna()].dropna(subset=["Height"])

gold = df_medalists[df_medalists["Medal"] == "Gold"]["Height"]
silver = df_medalists[df_medalists["Medal"] == "Silver"]["Height"]
bronze = df_medalists[df_medalists["Medal"] == "Bronze"]["Height"]

plt.figure(figsize=(8,6))
plt.boxplot([gold, silver, bronze], labels=["Gold", "Silver", "Bronze"])
plt.title("Distribución de Altura según Tipo de Medalla")
plt.ylabel("Altura (cm)")
plt.show()
```


- Las distribuciones de altura para los ganadores de medalla de oro, plata y bronce son muy similares. Las medianas se encuentran aproximadamente en el mismo rango, lo que indica que no existen diferencias significativas en la estatura promedio según el tipo de medalla obtenida.

- La variabilidad (rango intercuartílico) es parecida entre las tres categorías, lo que sugiere que la dispersión de alturas es consistente independientemente del tipo de medalla.

- Se observan valores atípicos tanto en alturas bajas como altas en todas las categorías, lo que refleja la diversidad física de los atletas que alcanzan el podio.

- En general, la altura no parece ser un factor determinante diferenciador entre quienes obtienen oro, plata o bronce, ya que las tres distribuciones presentan patrones muy similares.

5.2 Weight

```python
df_medalists = df[df["Medal"].notna()].dropna(subset=["Weight"])

gold = df_medalists[df_medalists["Medal"] == "Gold"]["Weight"]
silver = df_medalists[df_medalists["Medal"] == "Silver"]["Weight"]
bronze = df_medalists[df_medalists["Medal"] == "Bronze"]["Weight"]

plt.figure(figsize=(8,6))
plt.boxplot([gold, silver, bronze], labels=["Gold", "Silver", "Bronze"])
plt.title("Distribución de Peso según Tipo de Medalla")
plt.ylabel("Peso (kg)")
plt.show()
```


- Las distribuciones de peso para los ganadores de medalla de oro, plata y bronce son muy similares. Las medianas se encuentran en valores cercanos, lo que indica que no existen diferencias marcadas en el peso promedio según el tipo de medalla obtenida.

- La dispersión (rango intercuartílico) es comparable entre las tres categorías, lo que sugiere que la variabilidad del peso es consistente independientemente del tipo de medalla.

- Se observan numerosos valores atípicos en todas las categorías, especialmente en pesos altos. Esto refleja la diversidad de disciplinas deportivas incluidas en el conjunto de datos, ya que algunos deportes requieren mayor masa corporal que otros.

- En general, el peso, al igual que la altura, no parece ser un factor claramente diferenciador entre quienes obtienen oro, plata o bronce, dado que las tres distribuciones presentan patrones muy similares.
